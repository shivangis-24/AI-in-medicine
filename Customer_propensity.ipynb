Feature selection

```python
plt.figure(figsize = (20,15))
sns.heatmap(X.corr(),linewidth = 2)
plt.show()
```

```python
def correlation(dataset,threshold):
  correlated_set = set() #Set of all the correlated columns 
  corr_matrix = dataset.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range (i):
      if abs(corr_matrix.iloc[i,j]) > threshold:
        column_name = corr_matrix.columns[i]
        correlated_set.add(column_name)
  return correlated_set

      
```

```python
corr_features = correlation(X,0.7)
len(set(corr_features))
```

```python
corr_features 
```

```python
X = X.drop(corr_features,axis =1)
```

Model creation

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 42)
```

```python
print('X_train:', np.shape(X_train))
print('X_test:', np.shape(X_test))
print('Y_train:', np.shape(Y_train))
print('Y_test:', np.shape(Y_test))
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train,Y_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
Y_pred = classifier.predict(X_test)
```

```python
print(np.concatenate((Y_pred.reshape(len(Y_pred),1), Y_test.values.reshape(len(Y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
cm = confusion_matrix(Y_test,Y_pred)
print(cm)
```

```python
from mlxtend.plotting import plot_confusion_matrix
fig,ax = plot_confusion_matrix(conf_mat = cm, figsize = (6,6), cmap = plt.cm.Greens)
plt.xlabel('Predictions', fontsize = 18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()
```

```python
logreg = accuracy_score(Y_test,Y_pred)
rocreg = roc_auc_score(Y_test,Y_pred)
print('Accuracy:', logreg, '\nROC Score:', rocreg)
```

```python
import sklearn.metrics as metrics 
print(metrics.classification_report(Y_test,Y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
parameters_lr = [{'penalty':['l1','l2'],'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}]
grid_search_lr = GridSearchCV(estimator = classifier,
                              param_grid = parameters_lr,
                              scoring = 'accuracy',
                              cv = 10,
                              n_jobs = -1)
```

```python
grid_search_lr.fit(X_train, Y_train)
best_accuracy_lr = grid_search_lr.best_score_
best_parameter_lr = grid_search_lr.best_params_  
print("Best Accuracy of LR: {:.2f} %".format(best_accuracy_lr.mean()*100))
print("Best Parameter of LR:", best_parameter_lr)

```
